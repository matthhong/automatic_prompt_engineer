```python
import altair as alt
import pandas as pd
import numpy as np

# Generate multimodal data
np.random.seed(0)
data = np.concatenate([
    np.random.normal(0, 1, 1000),
    np.random.normal(5, 0.5, 500),
    np.random.normal(-3, 1, 800)
])

# Create a DataFrame
df = pd.DataFrame({'value': data})

# Create the histogram using Altair
bars = alt.Chart(df).mark_bar().encode(
    alt.X('value', bin=alt.Bin(step=1), title='Value'),
    alt.Y('count()', title='Frequency')
).properties(
    title='Multimodal Distribution Histogram',
    width=500,
    height=300
)

# Add an interaction layer
selection = alt.selection_multi(fields=['value'])
bars = bars.add_selection(selection)

tooltip = alt.Chart(df).mark_text(align='left', dx=5, dy=-5, color='black').encode(
    alt.X('value', bin=alt.Bin(step=1)),
    alt.Y('count()', stack='zero'),
    text='count()'
).transform_filter(selection)

# Combine the histogram and the tooltip layer
histogram = (bars + tooltip)

# Display the histogram
histogram
```
